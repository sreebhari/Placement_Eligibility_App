```python
import pymysql

#from db_config import get_connection

def create_tables():
    myconnection = get_connection()
    with myconnection:
        cur = myconnection.cursor()
        
        cur.execute("CREATE DATABASE IF NOT EXISTS placement_eligibiliy_db")
        cur.execute("USE placement_eligibiliy_db")
        
        # SQL statement to create the students table
        cur.execute("""
        CREATE TABLE students (
            student_id INT AUTO_INCREMENT PRIMARY KEY,
            name VARCHAR(100),
            age INT,
            gender VARCHAR(10),
            email VARCHAR(100),
            phone VARCHAR(20),
            enrollment_year INT,
            course_batch VARCHAR(50),
            city VARCHAR(50),
            graduation_year INT
        )""")

        #Altering table where 
        cur.execute("""ALTER TABLE students AUTO_INCREMENT = 1001""")


        # SQL statement to create the programming table
        cur.execute("""
        CREATE TABLE programming (
            programming_id INT AUTO_INCREMENT PRIMARY KEY,
            student_id INT,
            language VARCHAR(50),
            problems_solved INT,
            assessments_completed INT,
            mini_projects INT,
            certifications_earned INT,
            latest_project_score FLOAT,
            FOREIGN KEY (student_id) REFERENCES students(student_id)
        )""")
        
        #Altering table where 
        cur.execute("""ALTER TABLE programming AUTO_INCREMENT = 2001""")
        
        # SQL statement to create the soft_skills table
        cur.execute("""
        CREATE TABLE soft_skills (
            soft_skill_id INT AUTO_INCREMENT PRIMARY KEY,
            student_id INT,
            communication INT,
            teamwork INT,
            presentation INT,
            leadership INT,
            critical_thinking INT,
            interpersonal_skills INT,
            FOREIGN KEY (student_id) REFERENCES students(student_id)
        )""")
        
        #Altering table where 
        cur.execute("""ALTER TABLE soft_skills AUTO_INCREMENT = 3001""")
        
        # SQL statement to create the placements table
        cur.execute("""
        CREATE TABLE placements (
            placement_id INT AUTO_INCREMENT PRIMARY KEY,
            student_id INT,
            mock_interview_score INT,
            internships_completed INT,
            placement_status VARCHAR(20),
            company_name VARCHAR(100),
            placement_package FLOAT,
            interview_rounds_cleared INT,
            placement_date DATE,
            FOREIGN KEY (student_id) REFERENCES students(student_id)
        )""")
        
        #Altering table where 
        cur.execute("""ALTER TABLE placements AUTO_INCREMENT = 4001""")
        
        print("Tables created successfully.")

create_tables()
```

```python
import pymysql
from faker import Faker
import random
from datetime import datetime, timedelta
import pandas as pd
```

```python
fake = Faker()
```

```python
class Student:
    def __init__(self):
        self.name = fake.name()
        self.age = random.randint(21, 25)
        self.gender = random.choice(["Male", "Female"])
        self.email = fake.email()
        self.phone = str(random.randint(6000000000, 9999999999))
        self.enrollment_year = random.randint(2024, 2025)
        self.course_batch = random.choice(["Batch A", "Batch B", "Batch C"])
        self.city = fake.city()
        self.graduation_year = random.randint(2020, 2023)

    def insert(self, cur):
        insert_sql="""
        insert into students (name, age, gender, email, phone, enrollment_year, course_batch, city, graduation_year) values(%s, %s, %s, %s, %s, %s, %s, %s, %s)
        """
        data = (self.name, self.age, self.gender, self.email, self.phone, self.enrollment_year, self.course_batch, self.city, self.graduation_year)
        cur.execute(insert_sql, data)
        return cur.lastrowid

    
class Programming:
    def __init__(self, student_id):
        self.student_id = student_id
        self.language = random.choice(["Python", "SQL", "JavaScript"])
        self.problems_solved = random.randint(0, 100)
        self.assessments_completed = random.randint(0, 10)
        self.mini_projects = random.randint(0, 5)
        self.certifications_earned = random.randint(0, 3)
        self.latest_project_score = round(random.uniform(60, 100), 2)

    def insert(self, cur):
        sql = """
        INSERT INTO programming (student_id, language, problems_solved, assessments_completed, mini_projects,
        certifications_earned, latest_project_score)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        """
        data = (self.student_id, self.language, self.problems_solved, self.assessments_completed,
                self.mini_projects, self.certifications_earned, self.latest_project_score)
        cur.execute(sql, data)
        
        
        
class SoftSkills:
    def __init__(self, student_id):
        self.student_id = student_id
        self.communication = random.randint(50, 100)
        self.teamwork = random.randint(50, 100)
        self.presentation = random.randint(50, 100)
        self.leadership = random.randint(50, 100)
        self.critical_thinking = random.randint(50, 100)
        self.interpersonal_skills = random.randint(50, 100)

    def insert(self, cur):
        sql = """
        INSERT INTO soft_skills (student_id, communication, teamwork, presentation, leadership,
        critical_thinking, interpersonal_skills)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        """
        data = (self.student_id, self.communication, self.teamwork, self.presentation,
                self.leadership, self.critical_thinking, self.interpersonal_skills)
        cur.execute(sql, data)
        
        
class Placement:
    def __init__(self, student_id):
        self.student_id = student_id
        self.mock_interview_score = random.randint(0, 100)
        self.internships_completed = random.randint(0, 3)
        self.placement_status = random.choice(["Ready", "Not Ready", "Placed"])
        self.company_name = fake.company() if self.placement_status == "Placed" else None
        self.placement_package = round(random.uniform(3.0, 15.0), 2) if self.placement_status == "Placed" else None
        self.interview_rounds_cleared = random.randint(0, 5)
        self.placement_date = fake.date_between(start_date='-1y', end_date='today') if self.placement_status == "Placed" else None

    def insert(self, cur):
        sql = """
        INSERT INTO placements (student_id, mock_interview_score, internships_completed, placement_status,
        company_name, placement_package, interview_rounds_cleared, placement_date)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
        """
        data = (self.student_id, self.mock_interview_score, self.internships_completed, self.placement_status,
                self.company_name, self.placement_package, self.interview_rounds_cleared, self.placement_date)
        cur.execute(sql, data)
```

```python
def generate_data(n=100):
    connection = get_connection()
    with connection:
        cur = connection.cursor()
        for _ in range(n):
            student = Student()
            student_id = student.insert(cur)
    
            programming = Programming(student_id)
            programming.insert(cur)
    
            soft_skills = SoftSkills(student_id)
            soft_skills.insert(cur)
    
            placement = Placement(student_id)
            placement.insert(cur)

generate_data(500) 
print("500 student records inserted successfully.")
```

```python
#1. total number of students in each bach
pd.read_sql_query("""
select 
    course_batch
    ,count(*) as total_students 
from students group by 1 order by 1""",connection)
```

```python
#2. Top 5 Students Ready for Placement (Highest Mock Score)
pd.read_sql_query("""
SELECT 
    s.name
    , pl.mock_interview_score
    , p.problems_solved
FROM students s
JOIN placements pl ON s.student_id = pl.student_id
JOIN programming p ON s.student_id = p.student_id
JOIN soft_skills ss ON s.student_id = ss.student_id
WHERE pl.placement_status = 'Ready'
ORDER BY pl.mock_interview_score DESC
LIMIT 5;""",connection)
```

```python
#3. Average Programming Performance Per Batch
pd.read_sql_query("""
SELECT 
    s.course_batch
    ,p.language
    , round(AVG(p.problems_solved),2) AS avg_problems
FROM students s
JOIN programming p ON s.student_id = p.student_id
GROUP BY 1,2
ORDER BY 1,3 desc;""",connection)

```

```python
#4. Distribution of Soft Skills Scores (Average per Student)
pd.read_sql_query("""
SELECT 
    s.name,
    ROUND((ss.communication + ss.teamwork + ss.presentation + ss.leadership + ss.critical_thinking + ss.interpersonal_skills)/6, 2) AS avg_soft_skills
FROM students s
JOIN soft_skills ss ON s.student_id = ss.student_id
ORDER BY avg_soft_skills DESC;""",connection)
```

```python
#5. plecement percentage across each batch
pd.read_sql_query("""
SELECT 
    s.course_batch,
    COUNT(*) AS total_students,
    ROUND(
        (COUNT(CASE WHEN p.placement_status = 'Placed' THEN 1 END) / COUNT(*)) * 100, 0
    ) AS placed_percentage
    ,ROUND(
        (COUNT(CASE WHEN p.placement_status = 'Ready' THEN 1 END) / COUNT(*)) * 100, 0
    ) AS ready_percentage
    ,ROUND(
        (COUNT(CASE WHEN p.placement_status = 'Not Ready' THEN 1 END) / COUNT(*)) * 100, 0
    ) AS not_ready_percentage
FROM students s
JOIN placements p ON s.student_id = p.student_id
GROUP BY 1
ORDER BY 1;""",connection)
```

```python
#6. Placement Readiness vs Actual Placement
pd.read_sql_query("""
SELECT 
    placement_status
    , COUNT(*) AS count
FROM placements
WHERE placement_status IN ('Ready', 'Placed')
GROUP BY placement_status;
""",connection)
```

```python
#7. Students with Highest Number of Certifications

pd.read_sql_query("""
SELECT 
    s.name
    , p.certifications_earned
FROM students s
JOIN programming p ON s.student_id = p.student_id
ORDER BY p.certifications_earned DESC;
""",connection)
```

```python
#8. Average Placement Package by City

pd.read_sql_query("""
SELECT 
    s.city
    , AVG(pl.placement_package) AS avg_package
FROM students s
JOIN placements pl ON s.student_id = pl.student_id
WHERE pl.placement_status = 'Placed'
GROUP BY s.city
ORDER BY avg_package DESC;
""",connection)
```

```python
#9. top 10 Student with higest package

pd.read_sql_query("""
SELECT 
    s.name
    , pl.placement_package
FROM students s
JOIN placements pl ON s.student_id = pl.student_id
WHERE pl.placement_status = 'Placed'
ORDER BY placement_package DESC
limit 10;
""",connection)
```

```python
#10. Students with Highest Overall Score

pd.read_sql_query("""
SELECT s.name,
       p.problems_solved,
       ROUND((ss.communication + ss.teamwork + ss.presentation + ss.leadership + ss.critical_thinking + ss.interpersonal_skills)/6, 2) AS avg_soft_skills,
       (p.problems_solved + ((ss.communication + ss.teamwork + ss.presentation + ss.leadership + ss.critical_thinking + ss.interpersonal_skills)/6)) AS total_score
FROM students s
JOIN programming p ON s.student_id = p.student_id
JOIN soft_skills ss ON s.student_id = ss.student_id
ORDER BY total_score DESC
LIMIT 5;
""",connection)
```
